## Density-based spatial clustering of apllications with noise

### Idea of the algorithm:

Divide given set into groups,finding the densest neighborhoods of a given radius. In this algorithm all points will be devided into 3 categories:
* root-point: a point is called root if at least the specified number of points (MinPts) falls into her epsilon neighborhood.
* boundary point: a point, which has neighborhs less then specified in MinPts into her epsilon neighborhood, but falls into epsilon neighborhood of some root-point.
* noise: all the rest points call noise.

### Algorithm

* Choose randomly one point from the given set.
* If in the $\varepsilon$-neighborhood of this point is less then MinPts, choose another point.  
* Else include all its neighbors in a cluster.
* Repeat previous steps untill all clusters will be found.
* Points are not included in the any clusters are noise.

### Initial parameters
DBSCAN requires setting two parameters: $\varepsilon$ and the minimum number of points that should form a dense region (MinPts)

### Example

```python
import numpy as np

import matplotlib.pyplot as plt
from seaborn import scatterplot as scatter
from matplotlib import cm
%matplotlib inline

from sklearn.cluster import KMeans, MiniBatchKMeans
from sklearn.datasets import make_moons 

x, y = make_moons(n_samples=200, noise=0.05, random_state=0) 
plt.scatter(x[:, 0], x[:, 1]) 
plt.show()
```

Compare two different types of clustering.

```python
km = KMeans(n_clusters = 2, random_state=0)
y_pred_km = km.fit_predict(x)
plt.scatter(x[:, 0], x[:, 1], c=y_pred_km)
plt.show;
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
y_pred_db = db.fit_predict(x)
plt.scatter(x[:, 0], x[:, 1], c=y_pred_db)
plt.show;
```

### Algorithm weaknesses

* dimension curse effect
* difficult to find good initial parameters
* converges in $O(n^2)$
